# Edtech's Courses Creation Data 
 
We are analysing the data from Edx in which about the creation of different courses.

```python
file3 = file2.groupby('subject')['subject'].count().reset_index(name='count').sort_values(['count'], ascending= False).head(10)

file3
```

# Studying The Enrollments(Edx)

This is to understand that what subjects are getting maximum enrollments 

```python
file4 = file2[['subject','n_enrolled']].dropna(axis = 0)


file4['n_enrolled'] = file4['n_enrolled'].str.replace(",", "")
file4['n_enrolled'] = file4['n_enrolled'].str.strip()
file4['n_enrolled'] = file4['n_enrolled'].astype(int)


file4 = file4[['subject','n_enrolled']].groupby('subject').sum()

file4 = file4.sort_values(by = 'n_enrolled', ascending = False).head(10)
file4.plot.bar()
```

# We can clearly see that according to the recent trends Computer Science and Business Management is getting the most enrollments, and the job vacancies are maximun that area as well.
